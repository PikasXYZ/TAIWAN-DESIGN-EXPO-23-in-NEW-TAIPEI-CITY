```python
#@title import
import os
import time
import itertools
import pandas as pd
from selenium import webdriver
from google.colab import drive
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
#@title Main

drive.mount('/content/drive') #link to Google Drive

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #Runs in the background and doesn't display a visible window
chrome_options.add_argument('--no-sandbox') #Disables the sandboxing in Chrome
chrome_options.add_argument('--disable-dev-shm-usage') #Overcome limitations in Docker containers that cause Chrome to fail in certain scenarios

wd = webdriver.Chrome(options=chrome_options) #create a webdrive
wd.get("https://event.designexpo.org.tw/?utm_source=FB&utm_medium=Post&utm_campaign=Launch")

#@markdown ##Results will be saved to here👇:
save_to = "/content/drive/MyDrive" #@param {type:"string"}
file_name = os.path.join(save_to,"DesignExpo2023.csv")
if os.path.exists(file_name):
  results = pd.read_csv(file_name)
else:
  results = pd.DataFrame(columns=[f"Q{i}" for i in range(1,13)]+["Result"]) #Q1~Q12

#@markdown #####⠀⠀
#@markdown ##How many sec to wait when little troubles occur?
to_wait = 36  #@param {type:"slider", min:1, max:50, step:1}

try:
  for turn, elem in enumerate(itertools.product([0, 1], repeat=12), start=1): #all posibility of tuples that are composed of 0 or 1 and the size == 12.
    if turn > len(results):
      timer = time.time() #set timer

      start = WebDriverWait(wd, to_wait).until(EC.element_to_be_clickable((By.CLASS_NAME, "kv-btn")))
      start.click()

      Go = WebDriverWait(wd, to_wait).until(EC.element_to_be_clickable((By.CLASS_NAME, "start-btn")))
      Go.click()

      for i in elem: #elem is a tuple composed of 0 or 1 and its size is 12.
          if i :
              downside_option = wd.find_elements(By.CLASS_NAME, "ans-item")[1]
              downside_option.click()
          done = wd.find_element(By.CLASS_NAME, "next-btn") #the btn to the next page
          done.click()

      name = WebDriverWait(wd, to_wait).until(EC.element_to_be_clickable((By.NAME, "uname")))
      name.send_keys('test')

      done = wd.find_element(By.CLASS_NAME, "mix-btn")
      done.click()

      result = WebDriverWait(wd, to_wait).until(EC.element_to_be_clickable((By.XPATH  , "/html/body/div/section[4]")))
      result = int(result.get_attribute("class").split("r")[2]) # f"result r{i}" for i in range(1,17) == 16 sorts of result

      results.loc[len(results)] = [i for i in elem] + [result]
      print(f"Process : {turn:4}/4096\tSelection : {elem}\tResult : {result:2}\t(Time spent : {time.time()-timer:5.2f}sec)")

      wd.refresh()
except:
  results.to_csv(file_name, index=False)
  print(f"{len(results)} results are saved.")
else:
  results.to_csv(file_name, index=False)
  print(f"All results are saved.")

wd.quit()
```
